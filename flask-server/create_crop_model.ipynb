```python
import numpy as np
import pandas as pd
import time
import json

import pickle
import torch
import torch.nn as nn
from torchvision import models, transforms
from torch.autograd import Variable
from PIL import Image

#image feature extract
def feature_extract(image_dir):
    vgg_model = models.vgg19(pretrained=True)
    new_classifier = torch.nn.Sequential(*list(vgg_model.children())[-1][:6])
    new_classifier.add_module("Linear output", torch.nn.Linear(4096, 256))
    vgg_model.classifier = new_classifier
    trans = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])
    im = Image.open(image_dir).convert('RGB')
    im = trans(im)
    im.unsqueeze_(dim=0)
    vgg_model = vgg_model.eval()
    feature_extracted = vgg_model(im).data.numpy().tolist()[0]
    feature_extracted = [float(i) for i in feature_extracted]
#     print(feature_extracted)
#     print("before: ",feature_extracted.shape)
#     feature_extracted_fitted = StandardScaler().fit_transform(feature_extracted)
#     print(feature_extracted_fitted.shape)
#     pca = PCA(n_components=20)
#     pca_reduced = pca.fit_transform(feature_extracted)
#     print("after: ",pca_reduced.shape)

    return feature_extracted
image_dir = './lacuna_fund_eotg_v1_source_perspective/'
label_dir = './lacuna_fund_eotg_v1_labels/'
num_runs = 10000
image_features=[]
label_features=[]
start_time = time.time()

for i in range(0, num_runs):
    #image
    image_file = image_dir + 'lacuna_fund_eotg_source_perspective_'+str(i)+'/thumbnail.jpg'
    image_features.append(feature_extract(image_file))
    #label
    label_file = open(label_dir + 'lacuna_fund_eotg_labels_' + str(i) +'/vector_labels.json')
    labels = json.load(label_file)['features'][0]['properties']
    current_lable_data = []
    
    current_lable_data.append(labels['crop_name']) if 'crop_name' in labels else current_lable_data.append('')
    current_lable_data.append(float(labels['expected_yield'])) if 'expected_yield' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['drought_probability'])) if 'drought_probability' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['drought_extent'])) if 'drought_extent' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['growth_sowing'])) if 'growth_sowing' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['growth_vegetative'])) if 'growth_vegetative' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['growth_flowering'])) if 'growth_flowering' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['growth_maturity'])) if 'growth_maturity' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['disturbance_none'])) if 'disturbance_none' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['disturbance_weeds'])) if 'disturbance_weeds' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['disturbance_drought'])) if 'disturbance_drought' in labels else current_lable_data.append(0)
    current_lable_data.append(float(labels['disturbance_nutrient_deficit'])) if 'disturbance_nutrient_deficit' in labels else current_lable_data.append(0)
    label_features.append(current_lable_data)
    if i%100 == 0: 
        print(i)
    label_file.close()
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score, classification_report
import pickle
from sklearn.naive_bayes import GaussianNB
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# pca reduction
label_feature = np.array([label_features[:,0]])
image_features = np.array(image_features)

pca = PCA(n_components=10)
pca_reduced = pca.fit_transform(image_features)

print(label_feature, label_feature.shape)
print(pca_reduced,pca_reduced.shape)
#merge two array
merged_features = np.concatenate((label_feature.T,pca_reduced), 1)
# label_columns = ['label','expected_yield','drought_probability','drought_extent','growth_sowing','growth_vegetative','growth_flowering','growth_maturity','disturbance_none', 'disturbance_weeds','disturbance_drought',  'disturbance_nutrient_deficit'] 
label_columns = ['label'] 

image_columns = ['image_'+str(i) for i in range(0, 10)]
merged_colums = label_columns + image_columns
print(merged_features.shape)

print(len(merged_colums))
df = pd.DataFrame(data = merged_features, columns=merged_colums)
df
```

```python
pca_model_file = open('pca_model.pkl', 'wb')
pickle.dump(pca, pca_model_file)
pca_model_file.close()
```

```python
label_feature_name = ['expected_yield','drought_probability', 'drought_extent','growth_sowing','growth_vegetative', 'growth_flowering', 'growth_maturity', 'disturbance_none','disturbance_weeds','disturbance_drought','disturbance_nutrient_deficit']
image_feature_name = ['image_'+str(i) for i in range(0,10)]
feature_name = image_feature_name
```

```python
#read data
features = df[feature_name]
target = df['label']
labels = df['label']
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size = 0.2)
```

```python
target.unique()
```

```python
#logic regression
lr = LogisticRegression(C=1.0, class_weight='balanced')
lr.fit(Xtrain, Ytrain)
lr_predictions = lr.predict(Xtest)
lr_accuracy = accuracy_score(Ytest, lr_predictions)
lr_recall = recall_score(Ytest, lr_predictions, average='weighted')
lr_f1 = f1_score(Ytest, lr_predictions, average='weighted')
lr_precision = precision_score(Ytest, lr_predictions, average='weighted')
print("accuracy: ",lr_accuracy)
print("recall: ",lr_recall)
print("f1: ",lr_f1)
print("precision: ",lr_precision)
print(classification_report(Ytest, lr_predictions))
```

```python
#svm

svm = SVC(gamma='scale')

svm.fit(Xtrain,Ytrain)

svm_predictions = svm.predict(Xtest)

svm_accuracy = accuracy_score(Ytest, svm_predictions)
svm_recall = recall_score(Ytest, svm_predictions, average='weighted')
svm_f1 = f1_score(Ytest, svm_predictions, average='weighted')
svm_precision = precision_score(Ytest, svm_predictions, average='weighted')

print("accuracy: ",svm_accuracy)
print("recall: ",svm_recall)
print("f1: ",svm_f1)
print("precision: ",svm_precision)
print(classification_report(Ytest, svm_predictions))
```

```python
# Gaussian Naive Bayes
gnb = GaussianNB()
gnd_predictions = gnb.fit(Xtrain, Ytrain).predict(Xtest)
gnd_accuracy = accuracy_score(Ytest, gnd_predictions)
gnd_recall = recall_score(Ytest, gnd_predictions, average='weighted')
gnd_f1 = f1_score(Ytest, gnd_predictions, average='weighted')
gnd_precision = precision_score(Ytest, gnd_predictions, average='weighted')

print("accuracy: ",gnd_accuracy)
print("recall: ",gnd_recall)
print("f1: ",gnd_f1)
print("precision: ",gnd_precision)
print(classification_report(Ytest, gnd_predictions))
```

```python
#save models
lr_model_file = open('lr_model.pkl', 'wb')
pickle.dump(lr, lr_model_file)
lr_model_file.close()
svm_model_file = open('svm_model.pkl', 'wb')
pickle.dump(svm, svm_model_file)
svm_model_file.close()
gnb_model_file = open('gnb_model.pkl', 'wb')
pickle.dump(gnb, gnb_model_file)
gnb_model_file.close()
```
